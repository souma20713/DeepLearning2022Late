## ライブラリの読み込み

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

### グラフに出力するデータの準備

```python
t=np.arange(0.0,2.0,0.01) #0.0から2.0まで0.001刻みでデータを生成する
s=1+np.sin(2*np.pi*t)
fig,ax =plt.subplots()
ax.plot(t,s)
#グラフaxにラベルを設定する
ax.set(xlabel="time(s)",ylabel="voltage(mV)",title="About as simple as it gets,folks")
ax.grid()
ax.plot(t,s)
plt.show()
```

##  グラフのラベルとグリッドを描画させる

```python
#グラフaxにラベルを設定する
ax.set(xlabel="time(s)",ylabel="voltage(mV)",title="About as simple as it gets,folks")
ax.grid()
```
